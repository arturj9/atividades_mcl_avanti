##### 1.Escreva uma função que receba uma lista de números e retorne outra lista com os números ímpares.

```python
def filtra_lista_impares(lista: list[int]) -> list[int]:
    """
    Retorna uma nova lista contendo apenas os números ímpares da
    lista de entrada.

    Parâmetros:
    - lista (list[int]): Uma lista de números inteiros.

    Retorna:
    - list[int]: Uma nova lista contendo apenas os números ímpares da
    lista de entrada.

    Exemplo:
    >>> filtra_lista_impares([1, 2, 3, 4, 5])
    [1, 3, 5]
    """
    return [numero for numero in lista if numero % 2 != 0]
```

##### 2.Escreva uma função que receba uma lista de números e retorne outra lista com os números primos presentes.

```python
def eh_primo(numero: int) -> bool:
    """
    Verifica se um número é primo.

    Parâmetros:
    - numero (int): O número a ser verificado.

    Retorna:
    - bool: True se o número for primo, False caso contrário.

    Observação:
    Um número é considerado primo se for maior que 1 e não tiver divisores além de 1 e ele mesmo.

    Exemplo:
    >>> eh_primo(5)
    True
    >>> eh_primo(4)
    False
    """
    if numero <= 1:
        return False
    i = 2
    while i * i <= numero:
        if numero % i == 0:
            return False
        i = i + 1
    return True
```

```python
def filtra_lista_primos(lista: list[int]) -> list[int]:
    """
    Retorna uma nova lista contendo apenas os números primos da
    lista de entrada.

    Parâmetros:
    - lista (list[int]): Uma lista de números inteiros.

    Retorna:
    - list[int]: Uma nova lista contendo apenas os números primos da
    lista de entrada.

    Observação:
    Utiliza a função eh_primo() para verificar se um número é primo.

    Exemplo:
    >>> filtra_lista_primos([1, 2, 3, 4, 5])
    [2, 3, 5]
    """
    return [numero for numero in lista if eh_primo(numero)]
```

##### 3.Escreva uma função que receba duas listas e retorne outra lista com os elementos que estão presentes em apenas uma das listas.

```python
def estah_na_lista(numero: int, lista: list[int]) -> bool:
    """
    Verifica se um número está presente em uma lista.

    Parâmetros:
    - numero (int): O número a ser verificado se está presente na lista.
    - lista (list[int]): A lista de números inteiros onde será realizada
    a verificação.

    Retorna:
    - bool: True se o número estiver presente na lista, False caso contrário.

    Exemplo:
    >>> estah_na_lista(3, [1, 2, 3, 4, 5])
    True
    >>> estah_na_lista(6, [1, 2, 3, 4, 5])
    False
    """
    for numero_lista in lista:
        if numero == numero_lista:
            return True
    return False
```

```python
def obter_lista_unica(lista1: list[int], lista2: list[int]) -> list[int]:
    """
    Retorna uma lista contendo os elementos únicos presentes na
    lista1 que não estão em lista2.

    Parâmetros:
    - lista1 (list[int]): A primeira lista de números inteiros.
    - lista2 (list[int]): A segunda lista de números inteiros.

    Retorna:
    - list[int]: Uma lista contendo os elementos únicos presentes
    em lista1 que não estão em lista2.

    Exemplo:
    >>> obter_lista_unica([1, 2, 3, 4], [3, 4, 5, 6])
    [1, 2]
    >>> obter_lista_unica([1, 2, 3], [4, 5, 6])
    [1, 2, 3]
    """
    lista_unica = []
    for numero in lista1:
        if not estah_na_lista(numero, lista2):
            lista_unica.append(numero)
    return lista_unica
```

##### 4.Dada uma lista de números inteiros, escreva uma função para encontrar o segundo maior valor na lista.

```python
def encontrar_segundo_maior_valor(lista: list[int]) -> int:
    """
    Encontra o segundo maior valor em uma lista de números inteiros.

    Parâmetros:
    - lista (list[int]): A lista de números inteiros.

    Retorna:
    - int: O segundo maior valor na lista.

    Observação:
    Se a lista tiver menos de dois elementos, retorna 0.

    Exemplo:
    >>> encontrar_segundo_maior_valor([1, 2, 3, 4, 5])
    4
    >>> encontrar_segundo_maior_valor([5, 4, 3, 2, 1])
    4
    """
    primeiro = 0
    segundo = 0
    for numero in lista:
        if numero > primeiro:
            segundo = primeiro
            primeiro = numero
        elif numero > segundo:
            segundo = numero
    return segundo
```

```python
print(lista_exemplo)
print(encontrar_segundo_maior_valor(lista_exemplo))
```

##### 5.Crie uma função que receba uma lista de tuplas, cada uma contendo o nome e a idade de uma pessoa, e retorne a lista ordenada pelo nome das pessoas em ordem alfabética.

```python
def ordernar_pessoas_nomes_alfabetica(lista_de_pessoas: list[tuple]):
    """
    Ordena uma lista de pessoas pelo nome em ordem alfabética.

    Parâmetros:
    - lista_de_pessoas (list[tuple]): Uma lista de tuplas onde cada tupla
    representa uma pessoa,
    e o primeiro elemento da tupla é o nome da pessoa.

    Retorna:
    - list[tuple]: A lista de pessoas ordenada pelo nome em ordem alfabética.

    Exemplo:
    >>> ordernar_pessoas_nomes_alfabetica([('Maria', 25), ('João', 30),
    ('Ana', 20)])
    [('Ana', 20), ('João', 30), ('Maria', 25)]
    """
    lista_de_pessoas.sort(key=lambda x: x[0])
    return lista_de_pessoas
```
